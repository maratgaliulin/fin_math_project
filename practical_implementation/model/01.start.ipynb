```python
import pandas as pd
```

```python
df = pd.read_csv('data/EURUSD_01-12-2022_09-12-2023.csv', sep=';', dtype=str)
```

```python
df.columns = ['Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Vol']
```

```python
df[['Open', 'High', 'Low', 'Close', 'Vol']] = df[['Open', 'High', 'Low', 'Close', 'Vol']].astype('float32')
```

```python
def date_time_index_maker(dfr:pd.DataFrame):
    return pd.to_datetime(dfr['Date'][4:] + '-' + dfr['Date'][2:4] + '-' + dfr['Date'][0:2] + ' ' + dfr['Time'][0:2] + ':' + dfr['Time'][2:4] + ':' + dfr['Time'][4:])
```

```python
df_test = df.head()
```

```python
df_test['dttime'] = df_test.apply(date_time_index_maker, axis=1)
```
